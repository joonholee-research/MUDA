```python
import random
from sklearn import datasets
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
rnd = random.randint(1, 100000)
# rnd = 56110
random.seed(rnd)
np.random.seed(rnd)
```

```python
X, y = datasets.make_moons(n_samples=500, noise=0.05)
X2, y2 = datasets.make_moons(n_samples=500, noise=0.05) 

X2_cx = np.mean(X2[:,0])
X2_cy = np.mean(X2[:,1])
X2_c = np.array([[X2_cx, X2_cy],]*X2.shape[0])
rotate30 = np.array([[0.866, 0.500],
                     [-0.500,0.866]])
X2 = X2 - X2_c
Xt = np.matmul(X2, rotate30)
Xt = Xt + X2_c
plt.scatter(X[:,0], X[:,1], c=y, cmap=matplotlib.colors.ListedColormap(['red','green']))
plt.scatter(Xt[:,0], Xt[:,1], c='blue')

ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)
plt.grid(True)
```

```python
np.save('X_s', X)
np.save('y_s', y)
np.save('X_t', Xt)
np.save('y_t', y2)
```
